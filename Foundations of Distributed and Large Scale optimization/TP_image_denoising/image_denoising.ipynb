```python
from matplotlib import image as mpimg
import matplotlib.pyplot as plt
import pywt
import numpy as np
np.random.seed(2024)
```

# Question 3

```python
# Load image
path = "/Users/augustincablant/Documents/GitHub/3A-class/Foundations of Distributed and Large Scale optimization/TP_image_denoising/florence.jpg"
image = mpimg.imread(path)
plt.imshow(image)
plt.axis('off')  
plt.show()
print(image.shape)
```

```python
# Add noise
noise = np.random.normal(0, 30, image.shape)
noisy_image = image + noise
noisy_image = np.clip(noisy_image, 0, 255)
noisy_image = noisy_image.astype(np.uint8)
plt.imshow(noisy_image)
plt.axis('off')  
plt.show()
```

```python
for std in [10, 100, 1000]:    
    gauss = std * np.random.randn(*image.shape)
    y = image + gauss
    y = np.clip(y, 0, 255)
    y = y.astype(np.uint8)
    plt.imshow(y)
    plt.axis('off')  
    plt.show()
```
